Los datos del campo "valor" estan almacenados como string, por lo tanto hay que aplicarle funciones de string antes de poder realizar cualquier operacion aritmética

```python
inmuebles.Valor[0].split()
```

Al aplicarle .str, lo que te devuelve es una serie de pandas pero cada dato es una lista con los elementos del split (sin pasarle parámetros te lo separa por el espacio). Con expand=True le decis que te lo pase como dataframe, en vez de como serie.

```python
valor = inmuebles.Valor.str.split(expand=True)
inmuebles['Moneda'] = valor[0]
inmuebles['Precio'] = valor[1]
inmuebles.sample(3)
```

Para convertir el valor que esta en str a numero, primero hay que eliminar los puntos. Esto lo hacemos con replace() en donde el primer parámetro es lo que querés reemplazar y el segundo con qué lo queres reemplazar. Regex=True es un parámetro que hay que colocar en las nuevas versiones de python para que no te salga el warning.

```python
inmuebles['Precio'] = inmuebles['Precio'].str.replace('.', '', regex=True)
```

```python
inmuebles[['Barrio', 'Precio']]
```

```python
inmuebles['Precio_millon'] = inmuebles.Precio.astype('float')/1000000
inmuebles.info()
```

#**AULA 4: Machine Learning**
Importamos las librerías para crear el modelo de ML elegido, en este caso el de regresión lineal para predecir los valores de los inmuebles con los "datos de entrenamiento". Importamos la librería para realizar la validación del modelo con los "datos de prueba".Tomamos como línea base una variable que esté ligeramente correlacionada con la variable de salida (COD_UPZ_GRUPO con Precio_millon). 
Con test_size=0.25 indicamos que genere el entrenamiento con el 75% de los datos y la prueba con el  25%. 

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
X = datos_ml[['COD_UPZ_GRUPO']]
y = datos_ml['Precio_millon']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=99)
```

```python
X_train
```

```python
X_test
```

Aplicamos el modelo de ML propiamente dicho:

```python
modelo = LinearRegression()
```

Lo entrenamos:

```python
modelo.fit(X_train, y_train)
```

Le pedimos al modelo que prediga con X_test a y_test. Después podemos comparar los datos predichos con los reales (y_train) para corroborar la eficacia del modelo.

```python
y_predict_test = modelo.predict(X_test)
```

Importamos las métricas que nos van a dar una idea de qué tan bien ajustado está nuestro modelo: El error medio absoluto es muy alto (349..) y el r2 muy cercano a 0 (0.02) por lo tanto no está bien ajustado, ya que buscamos un error mínimo y un r2 cercano a 1.

```python
from sklearn.metrics import mean_absolute_error, r2_score
baseline_mae = mean_absolute_error(y_test, y_predict_test)

baseline_r2 = r2_score(y_test, y_predict_test)
print(baseline_mae, baseline_r2)
```

Intentamos mejorar nuestro modelo agregando más atributos:

```python
X = datos_ml[['COD_UPZ_GRUPO','Habitaciones','Banos','CONJUNTO_CERRADO','SALARIO_ANUAL_MI','TIENE_ESCRITURA']] 

Y = datos_ml["Precio_millon"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 99)
modelo_1 = LinearRegression()
modelo_1.fit(X_train, y_train)
y_predict_test = modelo_1.predict(X_test)
y_predict_train = modelo_1.predict(X_train)
mae_test = mean_absolute_error(y_test, y_predict_test)
r2_test = r2_score(y_test, y_predict_test)
mae_train = mean_absolute_error(y_train, y_predict_train)
r2_train = r2_score(y_train, y_predict_train)
print(mae_test,r2_test)
print(mae_train,r2_train)
```

Probamos el modelo pasandole los datos de un caso hipotético: (COD_UPZ_GRUPO=816, HABITACIONES=3, BANOS=2, CONJUNTO_CERRADO=1, SALARIO_ANUAL_MI=50, TIENE_ESCRITURA=1) Un inmueble con esas características valdría 407 millones de pesos.
